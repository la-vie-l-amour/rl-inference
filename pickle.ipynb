```python
with open("./ppo_buffer/rewads_buffer", "rb") as f:
    ppo = np.array(pickle.load(f))
    
with open("./ppo_buffer/stds_buffer", "rb") as f:
    ppo_std = np.array(pickle.load(f))
```

```python
x = range(1,len(ppo)+1)
fig = plt.figure(figsize=(10, 7), layout='constrained')
plt.plot(x, ppo, label='PPO', linewidth = 1.3, color = "#1111FF")

plt.xlabel('iteration')
plt.ylabel('reward')
plt.title(f"ppo reward_mean:{np.mean(np.array(ppo))}")
# plt.xlim((1,len(ppo)+1))
# plt.xticks([1,100,200,300,400,500,600,700,800])
# plt.legend()
plt.fill_between(x,ppo-1*ppo_std, ppo+1*ppo_std, color = "#1111FF",alpha = 0.2)
plt.savefig('ppo.png')
```

```python
with open("./sac_buffer/rewads_buffer", "rb") as f:
    sac = np.array(pickle.load(f))
    
with open("./sac_buffer/stds_buffer", "rb") as f:
    sac_std = np.array(pickle.load(f))
```

```python
x = np.array(range(0,len(sac)))
fig = plt.figure(figsize=(10, 7), layout='constrained')
plt.plot(x, sac, label='sac', linewidth = 1.3, color = "#1FB256")

plt.xlabel('iteration')
plt.ylabel('reward')
plt.title(f"SAC reward_mean:{np.mean(np.array(sac))}")
# plt.xticks([0,200,400,600,800,1000], [1,20000,40000,60000,80000,100000])
# plt.legend()
plt.fill_between(x,sac-1*sac_std, sac+1*sac_std, color = "#1FB256",alpha = 0.2)
plt.savefig('sac.png')
```

```python

    
with open("./FEEF_buffer/rewads_buffer", "rb") as f:
    feef = np.array(pickle.load(f))
    
with open("./FEEF_buffer/stds_buffer", "rb") as f:
    feef_std = np.array(pickle.load(f))
```

```python
x = np.array(range(0,len(feef)))
fig = plt.figure(figsize=(10, 7), layout='constrained')
plt.plot(x, feef, label='feef', linewidth = 1.3, color = "#FF1010")

plt.xlabel('iteration')
plt.ylabel('reward')
plt.title(f"FEEF reward_mean:{np.mean(np.array(feef))}")
# plt.xticks([1,50,100,150,200,250,300,350,400])
# plt.legend()
plt.fill_between(x,feef-1*feef_std, feef+1*feef_std, color = "#FF1010",alpha = 0.2)
plt.savefig('feef.png')
```

```python
m = min(len(ppo),len(sac),len(feef))
ppo = ppo[:m]
ppo_std = ppo_std[:m]
sac = sac[:m]
sac_std = sac_std[:m]
feef = feef[:m]
feef_std = feef_std[:m]
x =  np.array(range(m))
plt.figure(figsize=(10, 7), layout='constrained')
plt.plot(x, ppo, label='PPO',color = "#1111FF")  # Plot some data on the (implicit) axes.
plt.fill_between(x,ppo-1*ppo_std, ppo+1*ppo_std, color = "#1111FF",alpha = 0.2)
plt.plot(x, sac, label='SAC', color = "#1FB256")  # etc.
plt.fill_between(x,sac-1*sac_std, sac+1*sac_std, color = "#1FB256",alpha = 0.2)
plt.plot(x, feef, label='FEEF', color = "#FF1010")  # etc.
plt.fill_between(x,feef-1*feef_std, feef+1*feef_std, color = "#FF1010",alpha = 0.2)
plt.xlabel('iteration')
plt.ylabel('reward')
plt.title("Compare Algorithms")
plt.legend()
plt.savefig('compare.png')
```
